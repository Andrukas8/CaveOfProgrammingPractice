```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st

import warnings
warnings.filterwarnings('ignore')

### Setting Styles of the Graphs
sns.set(rc={'figure.figsize':(15,8)})
sns.set_style('ticks')
plt.style.use("dark_background")
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
# Subplots:

plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```
